```python
#Try AI directly in your favorite apps  Use Gemini to generate drafts and refine content, plus get Gemini Pro with access to Google's next-gen AI for Rs 5,600.00 PKR 0 for 1 month
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
# The original code used sns.load_dataset, which is for built-in example datasets.
# We need to use pd.read_csv to load a local file.
iris = pd.read_csv(r'/content/drive/MyDrive/Colab Notebooks/csv_data_iris (1).csv')

# Display dataset structure
print("Dataset Shape:")
print(iris.shape)
print("\nDataset Columns:")
print(iris.columns)
print("\nFirst 5 Rows:")
display(iris.head())

# Basic visualizations
plt.figure(figsize=(15, 10))

# Scatter plot to analyze relationships between variables
plt.subplot(2, 2, 1)
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='class', data=iris)
plt.title('Sepal Length vs Sepal Width')

plt.subplot(2, 2, 2)
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='class', data=iris)
plt.title('Petal Length vs Petal Width')

# Histogram to examine data distribution
plt.subplot(2, 2, 3)
sns.histplot(data=iris, x='sepal length (cm)', hue='class', kde=True)
plt.title('Distribution of Sepal Length')

plt.subplot(2, 2, 4)
sns.histplot(data=iris, x='petal length (cm)', hue='class', kde=True)
plt.title('Distribution of Petal Length')

plt.tight_layout()
plt.show()

# Box plot to detect outliers and spread of values
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=iris, x='class', y='sepal length (cm)')
plt.title('Sepal Length by Species')

plt.subplot(1, 2, 2)
sns.boxplot(data=iris, x='class', y='petal length (cm)')
plt.title('Petal Length by Species')

plt.tight_layout()
plt.show()
```
